```python
severity_mapping = {
    # Severity 1
    "NON-CRIMINAL": 'Severity 1',
    "SUSPICIOUS OCCURRENCE": 'Severity 1',
    "MISSING PERSON": 'Severity 1',
    "RUNAWAY": 'Severity 1',
    "RECOVERED VEHICLE": 'Severity 1',

    # Severity 2
    "WARRANTS": 'Severity 2',
    "OTHER OFFENSES": 'Severity 2',
    "VANDALISM": 'Severity 2',
    "TRESPASS": 'Severity 2',
    "DISORDERLY CONDUCT": 'Severity 2',
    "BAD CHECKS": 'Severity 2',

    # Severity 3
    "LARCENY/THEFT": 'Severity 3',
    "VEHICLE THEFT": 'Severity 3',
    "FORGERY/COUNTERFEITING": 'Severity 3',
    "DRUG/NARCOTIC": 'Severity 3',
    "STOLEN PROPERTY": 'Severity 3',
    "FRAUD": 'Severity 3',
    "BRIBERY": 'Severity 3',
    "EMBEZZLEMENT": 'Severity 3',

    # Severity 4
    "ROBBERY": 'Severity 4',
    "WEAPON LAWS": 'Severity 4',
    "BURGLARY": 'Severity 4',
    "EXTORTION": 'Severity 4',

    # Severity 5
    "KIDNAPPING": 'Severity 5',
    "ARSON":'Severity 5'
}
data['Severity'] = data['Category'].map(severity_mapping)
```

I have decided to break time to hour and date to year month and day so that we can apply a filter later to get particular rows and which could also help us in providing as a feature for ML model

```python
data['Dates'] = pd.to_datetime(data['Dates'])
data['date'] = data['Dates'].dt.date
data['time'] = data['Dates'].dt.time
data = data.drop(columns=['Dates'])

data['time'] = pd.to_datetime(data['time'], format='%H:%M:%S').dt.time
data['hour'] = pd.to_datetime(data['time'], format='%H:%M:%S').dt.hour

data['year'] = pd.to_datetime(data['date']).dt.year
data['month'] = pd.to_datetime(data['date']).dt.month
data['day'] = pd.to_datetime(data['date']).dt.day

data.drop(columns=['date'],inplace=True)
data.drop(columns=['time'],inplace=True)
```

in the orignal dataset we find that values of latitude and longitude have been flipped thus giving us incorretc loation when searched so we flip them back, we can see that the city is san fransico

```python
data['latitude']=data['Longitude (X)']
data['longitude']=data['Latitude (Y)']
data.drop(columns=['Latitude (Y)','Longitude (X)'],inplace=True)
```

```python

matplotlib.use('Agg')  # Set non-interactive backend

# Function to create and save plots
def create_bar_graph(severity_level):
    # Filter and process data
    severity_crimes = data[data['Severity'] == severity_level]
    resolution_counts = severity_crimes['Resolution'].value_counts().head(10).reset_index()
    resolution_counts.columns = ['Resolution', 'Count']

    # Create plot
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.barplot(x='Count', y='Resolution', data=resolution_counts, palette='viridis', ax=ax)
    ax.set_title(f'Top 10 Resolutions for {severity_level} Crimes')

    # Save and return
    #fig.savefig(f"{severity_level.lower().replace(' ', '_')}_resolutions.svg",
    #            format='svg', bbox_inches='tight')


    if severity_level == "Severity 1":
        fig.savefig("Res_for severity1.svg", format='svg', bbox_inches='tight')
    elif severity_level == "Severity 2":
        fig.savefig("Res_for severity2.svg", format='svg', bbox_inches='tight')
    elif severity_level == "Severity 3":
        fig.savefig("Res_for severity3.svg", format='svg', bbox_inches='tight')
    elif severity_level == "Severity 4":
        fig.savefig("Res_for severity4.svg", format='svg', bbox_inches='tight')
    elif severity_level == "Severity 5":
        fig.savefig("Res_for severity5.svg", format='svg', bbox_inches='tight')
    plt.close(fig)
    return fig

# Create tabs
severity_levels = ["Severity 1", "Severity 2", "Severity 3", "Severity 4", "Severity 5"]
tabs = widgets.Tab()

# Assign figures to tabs
tab_contents = []
for i, severity in enumerate(severity_levels):
    out = widgets.Output()
    with out:
        fig = create_bar_graph(severity)
        display(fig)  # Display the returned figure
    tab_contents.append(out)

tabs.children = tab_contents
for i, severity in enumerate(severity_levels):
    tabs.set_title(i, severity)

display(tabs)
```

#Trying to reduce the number of samples and testing it out [Balancing the classes]

as we know that our dataset is higly imbalance with certain classes having 100k examples while other having 200.

so we will try to balance it with undersmapling the majority class, adjusting the middle class and oversampling with slight variation in the smaller class

```python
LARGE_CLASS_SIZE = 5000
MID_SIZE_CLASS_SIZE = 4000
SMALL_CLASS_SIZE = 1000
```

```python
def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return list(synonyms)

# Function to add variations to text
def add_variations(text):
    words = text.split()

    # Synonym replacement
    if random.random() < 0.4:  # 30% chance to replace a word with a synonym
        for i, word in enumerate(words):
            synonyms = get_synonyms(word)
            if synonyms:
                words[i] = random.choice(synonyms)

    # Shuffle words
    if random.random() < 0.2:  # 20% chance to shuffle words
        random.shuffle(words)

    # Add prefixes/suffixes
    if random.random() < 0.2:  # 20% chance to add prefixes/suffixes
        prefixes = ['re', 'un', 'pre', 'mis']
        suffixes = ['ing', 'ed', 'ly', 'er']
        for i, word in enumerate(words):
            if random.random() < 0.1:
                words[i] = random.choice(prefixes) + word
            if random.random() < 0.1:
                words[i] = word + random.choice(suffixes)

    # Add random noise (typos)
    if random.random() < 0.1:  # 10% chance to add typos
        text_with_typos = []
        for char in ' '.join(words):
            if random.random() < 0.1:
                text_with_typos.append(random.choice('abcdefghijklmnopqrstuvwxyz'))
            else:
                text_with_typos.append(char)
        return ''.join(text_with_typos)

    return ' '.join(words)
```

Just from basic mapping we can see that there are a few outlier location, we can remove them using standard deviation method, any co ordinate very far from the mean location simply gets dropped

```python
def remove_location_outliers_std(df, lat_col='latitude', lon_col='longitude', n_std=3):
    """
    Removes outliers beyond `n_std` standard deviations from the mean.
    """
    lat_mean = df[lat_col].mean()
    lat_std = df[lat_col].std()

    lon_mean = df[lon_col].mean()
    lon_std = df[lon_col].std()

    filtered_df = df[
        (df[lat_col] >= lat_mean - n_std * lat_std) & (df[lat_col] <= lat_mean + n_std * lat_std) &
        (df[lon_col] >= lon_mean - n_std * lon_std) & (df[lon_col] <= lon_mean + n_std * lon_std)
    ]

    return filtered_df

# Example usage
df_clean1 = remove_location_outliers_std(southern)
df_clean2 = remove_location_outliers_std(traval)
df_clean3 = remove_location_outliers_std(tenderloin)
df_clean4 = remove_location_outliers_std(richmond)

df_clean5 = remove_location_outliers_std(park)
df_clean6 = remove_location_outliers_std(northern)
df_clean7 = remove_location_outliers_std(mission)
df_clean8 = remove_location_outliers_std(ingleside)

df_clean9 = remove_location_outliers_std(central)
df_clean10 = remove_location_outliers_std(bayview)
```
